```python
"""
389. Find the Difference
You are given two strings s and t. String t is generated by random shuffling string s and then add one 
more letter at a random position. Return the letter that was added to t.
Input: s = "abcd", t = "abcde"             Output: "e"
Input: s = "", t = "y"                     Output: "y"

this is basically to scan string t for a char that is missing in string s.
you can: 
1) using every char from t to search in string s, anyone missing will be reported. the time 
complexity is n*n, which is bad ---> not going to work for 'a' vs 'aa'. In coding, always pay
attention to duplicate values in string or list. dict and set are different.
2)build a dict from string s, and for any char that is missing in the dict, it will be reported. 
time complexity is O(N)(search in a dict is O(1))   
3. convert both s and t to a set, set(t)-set(s) will yield the missing char. need to use " ".join(set) 
to return it as string ---> not going to work for 'a' vs 'aa'
"""

class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        """
        :type s: str
        :type t: str
        :rtyp
        """
        # build a dictionary from string s
        hashmap = {}
        for char in s:
            hashmap[char] = hashmap.get(char, 0) + 1
            
        # search dictionary for char from string t
        for char in t:
            if char not in hashmap:
                return char
            else:
                hashmap[char] -= 1
            
"""
Runtime: 47 ms, faster than 71.12% of Python3 online submissions for Find the Difference.
Memory Usage: 13.9 MB, less than 31.86% of Python3 online submissions for Find the Difference.
"""     
```
