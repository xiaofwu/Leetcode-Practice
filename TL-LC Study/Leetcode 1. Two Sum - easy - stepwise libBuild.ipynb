```python
# 2022-09-16
# not a sorted array. There is a solution. not use same element twice
# Input: nums = [2,7,11,15], target = 9         Output: [0,1]
# [3,2,4]        6   [1,2]
# [3,3] 6    [0, 1]
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        if not nums: return []
        # start to build a dict using enumerate(), for every element, check if target - element in the dict
        # if target-element in the dict, return index. otherwise, add element to the dict using v:i
        hashmap = {}
        for i, v in enumerate(nums):
            if target - v in hashmap:
                return [hashmap[target - v], i]
            else:
                hashmap[v] = i
```
