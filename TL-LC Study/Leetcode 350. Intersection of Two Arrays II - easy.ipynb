```python
"""
350. Intersection of Two Arrays II
Given two integer arrays nums1 and nums2, return an array of their intersection. Each element in the 
result must appear as many times as it shows in both arrays and you may return the result in any order.

Input: nums1 = [1,2,2,1], nums2 = [2,2]               Output: [2,2]
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]           Output: [4,9]

go back to our traditional list --> dictionary stratagy. First convert nums1 into a dictionary with typic 
value:#of appearance as pair in the dict. Next, check every elements from nums2 for its presence in the dict. 
if there, append to the final list and (#of appearance -1). append as long as (#of appearanc)e >0.
"""

class Solution:
    def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
        """
        :type nums1: List[int]
        :typ nums2: List[int]
        :rtype List[int]
        """
        
        rList = []
        hashmap = {}
        
        #build a dictionary from nums1
        for value in nums1:
            if value not in hashmap:
                hashmap[value] = 1
            else:
                hashmap[value] += 1
                
        #check every values form nums2 in this dictionary
        for value in nums2:
            if value in hashmap and hashmap[value] > 0:
                rList.append(value)
                hashmap[value] -= 1
        
        return rList
"""
Runtime: 71 ms, faster than 65.16% of Python3 online submissions for Intersection of Two Arrays II.
Memory Usage: 14 MB, less than 52.00% of Python3 online submissions for Intersection of Two Arrays II.
"""                        
```

```python
# 2022-09-16
# 350. Intersection of Two Arrays II
class Solution:
    def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        if not nums1 or not nums2: return []
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1
        
        res = []
        hashmap = {}
        
        #build a dictionary from nums1 and count the appearance of element
        for i in nums1:
            hashmap[i] = hashmap.get(i, 0) + 1
                
        #check every value form nums2 in the dictionary
        for value in nums2:
            if value in hashmap and hashmap[value] > 0:
                res.append(value)
                hashmap[value] -= 1
        
        return res
"""
Runtime: O(N) 
Space: O(N) 
""" 
```
