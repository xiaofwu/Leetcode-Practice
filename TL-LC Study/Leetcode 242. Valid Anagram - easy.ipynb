```python
"""
242. Valid Anagram
Given two strings s and t, return true if t is an anagram of s, and false otherwise.
An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically 
using all the original letters exactly once.
Input: s = "anagram", t = "nagaram"                 Output: true
Input: s = "rat", t = "car"                         Output: false

build a dictionary from the string s (value:count of appearance). use every char from t to check dict, if 
char in the dict, #of appearance of this char -1. In the end, if all values in the dict is 0, then two strings are 
anagram
"""

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        hashmap = {}
        
        #build a dict to count char appearance
        for char in s:
            if char not in hashmap:
                hashmap[char] = 1
            else:
                hashmap[char] += 1
        
        #check char from string t in the dictionary
        for char in t:
            if char not in hashmap:
                return False
            else:
                hashmap[char] -= 1
                
        #check if all values in the dicttionary are 0
        for key in hashmap:
            if hashmap[key] != 0:
                return False
        return True
"""
Runtime: 71 ms, faster than 60.31% of Python3 online submissions for Valid Anagram.
Memory Usage: 14.4 MB, less than 67.19% of Python3 online submissions for Valid Anagram.
"""                     
        
```
