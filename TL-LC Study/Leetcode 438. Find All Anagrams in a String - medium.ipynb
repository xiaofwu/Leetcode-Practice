```python
# 2022-09-28
# 438. Find All Anagrams in a String
# Input: s = "cbaebabacd", p = "abc"      Output: [0,6]

class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        """
        :type s: str
        :type p: str
        :rtype: List[int]
        """
        from collections import Counter
        
        # Corner cases check
        if not s or not p: return []
        if len(s) < len(p): return []
        elif len(s) == len(p):
            if Counter(s) == Counter(p): return [0]
            else: return []
        
        res = []
        dict_p = Counter(p)
        window = Counter(s[:len(p)-1])
        slow = 0
        
        # start from char in front of window
        for i in range(len(p)-1, len(s)):
            # add char s[i] into the small window
            window[s[i]] = window.get(s[i], 0) + 1
            
            # check if larger window is the anagram of p string   
            if window == dict_p:
                res.append(slow)
            
            # after check, remove s[slow] from window
            # update slow to move the window
            window[s[slow]] -= 1
            if window[s[slow]] == 0:
                del window[s[slow]]
            slow += 1
        return res
"""
Time: O(n)
Space: O(len(p))
"""       
```
