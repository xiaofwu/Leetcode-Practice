```python
"""
169. Majority Element
Given an array nums of size n, return the majority element. The majority element is the element that appears 
more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.
Input: nums = [2,2,1,1,1,2,2]          Output: 2
Input: nums = [3,2,3]                  Output: 3
"""
#This is a typical list to dictionary conversion with value's appearance count as extra attribute

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        """
        :type nums: Lsist[int]
        :rtype int
        """
        
        newDict = {}
        if not nums:
            return None
        if len(nums) == 1:
            return nums[0]
        #build a new dict from input list with appearance count
        for value in nums:
            if value not in newDict:
                newDict[value] = 1
            else:
                newDict[value] += 1
                
        #check the dict, return value with appearcnce > n/2
        for value in newDict:
            if newDict[value] > len(nums)/2:
                return value
            
"""
Runtime: 245 ms, faster than 58.36% of Python3 online submissions for Majority Element.
Memory Usage: 15.3 MB, less than 98.92% of Python3 online submissions for Majority Element.
"""
```

```python
# 2022-09-16 摩尔投票法
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        """
        :type nums: Lsist[int]
        :rtype int
        """
        if not nums: return 0
        
        majority, count = 0, 0
        # assign first element to 
        for i in nums:
            if count == 0:
                majority = i
            if i == majority:
                count += 1
            else:
                count -= 1
        return majority

"""
time: O(N)
space: O(1)
"""       
```
