```python
"""
349. Intersection of Two Arrays
Given two integer arrays nums1 and nums2, return an array of their intersection. Each element 
in the result must be unique and you may return the result in any order.

Input: nums1 = [1,2,2,1], nums2 = [2,2]            Output: [2]
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]        Output: [9,4]
"""
class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        # intuitively, we can somehow use set to get overlap between two list
        # either unique or not. 
        return list(set(nums1)&set(nums2))    
"""
time O(n)
space O(n)
"""
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        # Actually more intuitively you will think about using a dictionary
        # build a dictonary using small nums(to save space), then loop through
        # elements in another list
        # it actually runs pretty fast!
        
        if not nums1 or not nums2: return []
        if len(nums1) > len(nums2): nums1, nums2 = nums2, nums1
        
        # build a dictionary from nums1
        res = []
        hashmap = {v:i for i, v in enumerate(nums1)}
        for i in set(nums2):
            if i in hashmap: res.append(i)
        return res
"""
time O(n)
space O(n)
"""        
```

```python
#22-09-18
# 349. Intersection of Two Arrays
# thi is BS version solution
class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        if not nums1 or not nums2: return []
        nums1, nums2 = list(set(nums1)), list(set(nums2))
        if len(nums1) > len(nums2): 
            nums1, nums2 = nums2, nums1
            
        #use element from nums1 to BS a sorted nums2
        res = []
        nums2.sort()
        for i in nums1: 
            L, R = 0, len(nums2) - 1
            while L <= R:
                mid = L + (R - L) // 2
                if nums2[mid] == i: 
                    res.append(i)
                    break
                elif nums2[mid] < i:
                    L = mid + 1
                else:
                    R = mid -1
        return res
"""
time O(nlogn)
space O(n)
"""                                
```
