```python
"""
229. Majority Element II
Given an integer array of size n, find all elements that appear more than ⌊ n/3 ⌋ times.
Input: nums = [3,2,3]             Output: [3]

Input: nums = [3,2,3]            Output: [3]
Input: nums = [1,2]              Output: [1,2]

Follow up: Could you solve the problem in linear time and in O(1) space?


this is quite easy to slove if we build a hashmap and count the appearance, append values oappearing over n/3
to a new list.

BUT HOW TO SOLVE IT IN SPACE? -- TWO OINTER? NO
摩尔投票算法
***重点： 因为选中的元素要大于 n/3 ---》这意味着我们最多能选到2个元素。 不然数量就超标了
1) let's initiate 4 variable: candidate1 candidate2, count1, count2
2) assign first two unique element to candiate 1 and 2. Assume they are the two over-respresented elements in the list
3) Way to initiate: candidate 1,2, count1,2 = 0. for n in... if count1 ==0, candidate 1=n; elif count2 ==0, 
    candidate 1=n
3) use for loop to go over all elements in the list and do: a) acount1+1 every time you see candidate1 and do same
    for candidate 2; b)acount1-1 AND account2-1 every time you see a different element. if 哪个count ==0, a new
    element will be assigned to 相应的 candidate，count 会重新开始计数。
4） 最后， 还要用for loop 重新go over the list, and count candidate1, and 2. Return the one (or both) if 
    count1,or2 > len(list) // 3  (note: a // b Floor division 向下取整除，地板除) （还有向上取整除，就是天花板除，
    ceil division，在math包中的 ceil() method

"""
```

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        if not nums:
            return []
        # initiate 4 variables for candidate 1&2 and their count 1&2
        cand1, cand2 = 0, 0
        count1, count2 = 0, 0
        res = []
        
        for n in nums:
            # use count1 to record cand1 appearance. it is important to do this first
            if cand1 == n:    
                count1 += 1
            elif cand2 == n:
                count2 += 1
                
            # assign first 2 unique elements to cand1 and cand2. 
            elif count1 == 0:  
                cand1 = n
                count1 =1
            elif count2 == 0:
                cand2 = n
                count2 = 1
                
            # decrease count if we see a different element in the list (different from cand 1 and 2)    
            else:            
                count1 -= 1
                count2 -= 1
                
        # after for loop, we still need to count cand1,2 again to make sure over 3/n
        count1, count2 = 0, 0 
        for n in nums:
            if cand1 == n:
                count1 += 1
            elif cand2 == n:
                count2 += 1
        
        if count1 > len(nums) // 3:
            res.append(cand1)
        if count2 > len(nums) // 3:
            res.append(cand2)
            
        return res
"""
O(N) Runtime (go through list twic)
O(1) Space

Runtime: 124 ms, faster than 90.95% of Python3 online submissions for Majority Element II.
Memory Usage: 15.4 MB, less than 6.92% of Python3 online submissions for Majority Element II.

"""               
```
