```python
"""
15. 3Sum
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that 
i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.
Notice that the solution set must not contain duplicate triplets.

Input: nums = [-1,0,1,2,-1,-4]        Output: [[-1,-1,2],[-1,0,1]]

Solution 1: 
1) most intuitive way: use double for loop to setup a dictionary {{i, j}: nums[i]+nums[j]}
2) for any nums[k], if nums[k] + dict[{i,j}] =0, and k not in {i, j}, then 
    rList.append([k] + list{i,j})
3） problem is how to deal with duplicate triplicate. I tried to use set() to convert all small lists in
    rList to small sets and use set to elminate duplicate and then convert small sets back to lists 
    --- > very hard to implement because " All built-in immutable objects, like tuples, are hashable while 
    mutable containers like lists and dictionaries are not hashable". That means list is not hashable, while
    set is harshable. That makes it difficult to convert between list and set
    
Solution from Youtube:
This is the solution based on Two Sum and Two Sum II, plus idea of using 3 pointers. This is a very good solution!!!
idea is as the following:
1) First, we need to sort the list in order to use two pointers to move in both directions.
2) Let's start from nums[0] (current pointer), then we set left pointer to nums[1], and right pointer to nums[len-1].
    Next we start to move two pointers toward each other. Every stop, we calculate sum = current+left+right. if sum=0,
    we append a list of three elements pointed to a rList. Keep moving while left < right.
3) How to deal with duplicate triplicate:
    a) when we move current pointers, we skip it and moves on if the next element has the same value. 
    b) we also need to do the same thing when we move left and right pointers. Every time we meet the same value, 
        we skip it and move pointer one more step.
                                            
                                            VERY NICE CODING!!!
                  先给数组排序，确定一个数，枚举另外两个数，注意去重，如果第一个数已经是正数了，则可以退出
                                            
"""
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        """
        :type nums: List[int]
        rtype: List[int]
        """
        # first step: sort the array
        # we need a sorted array in order to move left and right pointer based on sum
        nums.sort()  
        
        triplets = []
        
        # start to loop using three pointers
        for i in range(len(nums) -2):
        # if "current" moves and ponts to a same value, skip it.
            if i > 0 and nums[i] == nums[i - 1]:  
                continue  
            
            left = i + 1
            right = len(nums) -1
            # start to move left and right pointers for calculating sum while current pointer stays put.   
            while left < right: 
                currentSum = nums[i] + nums[left] + nums[right]
                
                if currentSum == 0:
                    triplets.append([nums[i], nums[left], nums[right]])
                    # check if left or right pointer moves to a same value
                    while left < right and nums[left] == nums[left + 1]:   
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                            
                    left += 1
                    right -= 1
                # condition check to decide which pointer to move    
                elif currentSum < 0:   
                    left += 1
                else:
                    right -= 1
            
        return triplets

#O(N^2) RunTime
#O(N) Space
                
"""
Runtime: 767 ms, faster than 91.21% of Python3 online submissions for 3Sum.
Memory Usage: 18.1 MB, less than 72.13% of Python3 online submissions for 3Sum.
"""                
                
            
        
        
```
