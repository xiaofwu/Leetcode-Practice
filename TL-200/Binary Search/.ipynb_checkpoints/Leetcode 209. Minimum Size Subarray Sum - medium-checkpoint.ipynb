209. Minimum Size Subarray Sum
Given an array of positive integers nums and a positive integer target, return the minimal length of a contiguous subarray [numsl, numsl+1, ..., numsr-1, numsr] of which the sum is greater than or equal to target. If there is no such subarray, return 0 instead.

Input: target = 7, nums = [2,3,1,2,4,3]                Output: 2
Explanation: The subarray [4,3] has the minimal length under the problem constraint.

Input: target = 11, nums = [1,1,1,1,1,1,1,1]           Output: 0

Input: target = 4, nums = [1,4,4]
Output: 1

```python
# Python 双指针, 使用滑动窗口和指针计算符合条件的最短子序列. 时间复杂度：O(n)    空间复杂度：O(1)
# this is the best solution for this question
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        # cover corner cases
        if not nums: return 0
        if sum(nums) < target: return 0

        minLen = float('-inf')
        total = 0
        j = 0
        
        # use a sliding window to record all possible subarrays 
        # that sum >= target for its window size
        for i in range(len(nums)): 
            # keep adding element into the window till total > target
            while j < len(nums) and total < target:
                total += nums[j]
                j += 1
            # when total>target, update res if possible. pop nums[i]. then go back to for loop
            if total >= target and j - i < res:
                res = j - i 
            total -= nums[i]
        return res if res < (len(nums) + 1) else 0
```
