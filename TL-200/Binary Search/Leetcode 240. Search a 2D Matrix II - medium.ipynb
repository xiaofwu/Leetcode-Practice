240. Search a 2D Matrix II
Write an efficient algorithm that searches for a value target in an m x n integer matrix matrix. This matrix has the following properties:

<br>Integers in each row are sorted in ascending from left to right.
<br>Integers in each column are sorted in ascending from top to bottom.
<br>m == matrix.length row 
<br>n == matrix[i].length column 

<br>Input: matrix = [[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]], target = 5
<br>Output: true

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if len(matrix) == 0: return False
        # we initialize the pointer at the bottom-left of the matrix
        r, c = len(matrix)-1, 0
        while r >=0 and c <len(matrix[0]):
            if matrix[r][c] == target:
                return True
            else:
                if matrix[r][c] < target:
                    c += 1
                else:
                    r -= 1
        return False
```

Time complexity: O(m+n)
m is the number of rows, and n is the number of columns.
The worst-case here is that the algorithm needs to finish scan one row and one column.
Space complexity: O(1)
The algorithm only needs four integers : row, col, height, width, so the space complexity is constant.

Since all the rows of the matrix are sorted, it would make sense to go through all the rows and look for the target value using binary search. This is also a good problem to solve if you are looking to find an example application of binary search.

The time complexity of binary search is O(logn), and since we perform binary search on all the rows, the final time complexity of the whole solution is O(nlogn).

```python
# 
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        
        def binSearch(nums, target):

            lo, hi = 0, len(nums)

            while lo < hi:
                mid = (lo + hi) // 2
                if nums[mid] == target:
                    return True
                elif nums[mid] < target:
                    lo = mid + 1
                elif nums[mid] > target:
                    hi = mid

            return False
        
        for row in matrix:
            if binSearch(row, target):
                return True 
            
        return False
```
