```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        stack = []
        res = []
        
        def bt(openN,closeN):
            if openN == closeN == n:
                res.append("".join(stack))
                return 
            
            if openN < n: 
                stack.append("(")
                bt(openN + 1, closeN)
                stack.pop()
            
            if closeN < openN:
                stack.append(")")
                bt(openN, closeN + 1)
                stack.pop()
        bt(0,0)
        
        return res
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        
        if not nums: return 0
        if len(nums) == 1: return nums[0]
        if len(nums) == 2: return max(nums[0],nums[1])
        
        dp = [0]*len(nums)
        for i in range(2,len(nums)):
            dp[0] = nums[0]
            dp[1] = max(nums[0],nums[1])
            dp[i] = max(dp[i-2] + nums[i], dp[i-1])
        return dp[-1]
    
    
        #you rob this house... dp[i-2] + nums[i]
        #you didn't rob this house dp[i-1]
#T:O(N)
#S:O(N)
```

```python
11. 213. House Robber II
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        
        #you can rob first one to -2 one
        #you can rob second one to last one 
        
        def max_rob(x):
            
            if not x: return 0
            if len(x) == 1: return x[0]
            if len(x) == 2: return max(x[0],x[1])
            
            if len(x) > 2:
                
                dp0 = x[0]
                dp1 = max(x[0], x[1]) 
                
                for i in range(2,len(x)):
                    dp = max(x[i] + dp0,dp1)
                    dp0,dp1 = dp1,dp
                return dp
            
        if len(nums) == 1: return nums[0]
        
        return max(max_rob(nums[0:-1]), max_rob(nums[1:]))
#T:O(N)
#S:O(1)
```

```python
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        
        dp = [0] * len(nums)
        dp[0] = nums[0]
        for i in range(1,len(nums)):
            dp[i] = dp[i-1] + nums[i]
        return dp
#T:O(N)
#S:O(N)
```

```python
13. 1143. Longest Common Subsequence
```

```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """
        ### this problem is simiar to problem 62, Unique path. 

        m = len(text1)+1
        n = len(text2)+1

        dp = [[0]*n for _ in range(m)]
        #n column, m row
        
        for i in range(1,m):
            for j in range(1,n):
                #calculate row by row 
                if text1[i-1] == text2[j-1]:
                    dp[i][j] = dp[i-1][j-1]+1
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
        return dp[-1][-1]
    #T: O(M*N)
    #S:O(M*N)
```

```python
class Solution(object):
    def LCS(self, word1, word2):
        
        array = [[0]*(len(word2)+1) for i in range(len(word1)+1)]

        for m in range(len(word1)):
            for n in range(len(word2)):
                if word1[m] == word2[n]:
                    array[m+1][n+1] = array[m][n] + 1
                else:
                    array[m+1][n+1] = max(array[m][n+1], array[m+1][n])
        return array[-1][-1]
    
    
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """
        #This is just a LCS problem. Please check solution 1143.

        return len(word1)+len(word2)-2*self.LCS(word1, word2)
```
