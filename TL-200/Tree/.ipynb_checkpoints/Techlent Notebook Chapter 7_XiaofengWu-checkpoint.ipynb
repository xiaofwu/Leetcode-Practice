### <font color='red'>102. Binary Tree Level Order Traversal</font>


Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).
<div></div> 
Input: root = [3,9,20,null,null,15,7]        Output: [[3],[9,20],[15,7]]

```python
# a typical BFS problem. Easier to implement using queue
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root: return []
        
        # BFS using queue is similar to DFS using stack
        # visit nodes in the same level, from top to bttom
        # use for loop of queue's size to make sure to pop all nodes in one level
        import collections
        queue = collections.deque([root])
        res = []
        
        while queue:
            cur_level = []
            size = len(queue)
            
            for i in range(size):
                check = queue.popleft()
                cur_level.append(check.val)
                if check.left: 
                    queue.append(check.left)
                if check.right:
                    queue.append(check.right)
            res.append(cur_level)
        
        return res
```

```python
# Recursive Pre-Order Traversal
# This is not very intuitive compared to BFS using queue.
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        res = []
        
        def dfs(node, depth):
            if not node: return
            
            if depth >= len(res):
                res.append([])
                
            res[depth].append(node.val)
            dfs(node.left, depth + 1)
            dfs(node.right, depth + 1)
        
        dfs(root, 0)
        return res
```

### <font color='red'>94. Binary Tree Inorder Traversal</font>


Given the root of a binary tree, return the inorder traversal of its nodes' values.
<div></div> 
Input: root = [1,null,2,3]            Output: [1,3,2]

```python
# implement inorder using stack
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        # recursive
        if not root: return []
        
        res = []
        stack = []
        
        while stack or root:
            if root:
                stack.append(root)
                root = root.left
            else:
                tmp = stack.pop()
                res.append(tmp.val)
                root = tmp.right
        return res
```

```python
### recursive version
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        # recursive
        if not root: return []
        
        return self.inorderTraversal(root.left)+[root.val]+self.inorderTraversal(root.right)
```

```python
### dfs version
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root: return []
        
        res = []
        self.dfs(root, res)
        return res
    
    def dfs(self, root, lst):
        if not root: return
        
        if root.left:
            self.dfs(root.left, lst)
        lst.append(root.val)
        if root.right:
            self.dfs(root.right, lst)
```

### <font color='red'>100. Same Tree</font>


Given the roots of two binary trees p and q, write a function to check if they are the same or not.
Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.
<div></div> 

Input: p = [1,2,1], q = [1,1,2]      Output: false

```python
# from Techlent notebook
class Solution(object):
        def isSameTree(self, p, q):
            """
            :type p: TreeNode
            :type q: TreeNode
            :rtype: bool
            """
            if p and not q:
                return False
            if not p and q:
                return False
            if not p and not q:
                return True            
            
            stack = [p, q]
            
            while stack:
                p = stack.pop()
                q = stack.pop()
                if p.val != q.val:
                    return False
                
                if p.left and not q.left:
                    return False
                if q.left and not p.left:
                    return False
                if p.left and q.left:
                    stack.append(p.left)
                    stack.append(q.left)
                
                if p.right and not q.right:
                    return False
                if q.right and not p.right:
                    return False
                if p.right and q.right:
                    stack.append(p.right)
                    stack.append(q.right)
            return True
    
```

```python
# BFS with queue    
def isSameTree3(self, p, q):
    import collections
    queue = collections.deque([(p, q)])

    while queue:
        node1, node2 = queue.popleft()
        if not node1 and not node2:
            continue
        elif not node1 or not node2::
            return False
        else:
            if node1.val != node2.val:
                return False
            queue.append((node1.left, node2.left))
            queue.append((node1.right, node2.right))
    return True
```

```python
# DFS with stack        
def isSameTree2(self, p, q):
    stack = [(p, q)]
    while stack:
        node1, node2 = stack.pop()
        if not node1 and not node2:
            continue
        elif None in [node1, node2]:
            return False
        else:
            if node1.val != node2.val:
                return False
            stack.append((node1.right, node2.right))
            stack.append((node1.left, node2.left))
    return True
```

```python
# recursive version
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q: return True
        if not p or not q: return False
        
       
        if p.val != q.val:
            return False
        else:
            return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
```

### <font color='red'>101. Symmetric Tree
</font>


Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).
<div></div> 

Input: root = [1,2,2,3,4,4,3]           Output: true

```python
# We can do this using either stack or queue
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root: return True
        stack = [(root.left, root.right)]
        while stack:
            left, right = stack.pop()
            if not left and not right:
                continue
            if not left or not right:
                return False
            if left.val == right.val:
                stack.append((left.left, right.right))
                stack.append((left.right, right.left))
            else:
                return False
        return True
```

```python
# recursive version
class Solution(object):
    def isSymmetric(self, root):
        if not root:
            return True
        
        def dfs(left,right):
            # 
            # 
            # 
            if not (left or right):
                return True
            if not (left and right):
                return False
            if left.val!=right.val:
                return False
            return dfs(left.left,right.right) and dfs(left.right,right.left)
        
        return dfs(root.left,root.right)
```

```python
# or code it this way
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        
        return self.dfs(root.left,root.right)
    def dfs(self, left,right):
        # 
        # 
        # 
        if not (left or right):
            return True
        if not (left and right):
            return False
        if left.val!=right.val:
            return False
        return self.dfs(left.left,right.right) and self.dfs(left.right,right.left)
```

```python
# recursive version. a little bit complicated
class Solution:
    def isSymmetric(self, root):
        if not root: return True
        else:
            return self.isMirror(root.left, root.right)

    def isMirror(self, left, right):
        if not left and not right:
            return True
        if not left or not right:
            return False

        if left.val == right.val:
            outPair = self.isMirror(left.left, right.right)
            inPiar = self.isMirror(left.right, right.left)
            return outPair and inPiar
        else:
            return False
```

### <font color='red'>104. Maximum Depth of Binary Tree</font>


Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.
<div></div> 
Input: root = [3,9,20,null,null,15,7]            Output: 3

```python
# recursive version
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        
        left = self.maxDepth(root.left)
        right = self.maxDepth(root.right)
        
        return max(left, right) + 1
```

```python
#This is the queue version.
#It is the level order or BFS way to check each level
# we can also use a stack to replace queue. code is the same.
class Solution(object):
        def maxDepth(self, root):
            if not root:
                return 0
            
            level = 1
            max_depth = 0
    
            from collections import deque
            queue = deque()
            queue.append((root, level))
            
            while queue:
                check, level = queue.popleft()
                max_depth = max(max_depth, level)
                if check.left:
                    queue.append((check.left, level+1))
                if check.right:
                    queue.append((check.right, level+1))
            return max_depth
```

### <font color='red'>144. Binary Tree Preorder Traversal</font>


Given the root of a binary tree, return the preorder traversal of its nodes' values.

```python
# same as Techlent notebook. use stack for preorder
# note: check right child first 
class Solution(object):
    def preorderTraversal(self, root):
        if not root:
            return []
        
        stack = [root]
        rL = []
        while stack:
            check = stack.pop()
            rL.append(check.val)
            if check.right:
                stack.append(check.right)
            if check.left:
                stack.append(check.left)
        return rL
```

```python
# recursive version
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        return [root.val] + self.preorderTraversal(root.left) + self.preorderTraversal(root.right)
```

### <font color='red'> 145. Binary Tree Postorder Traversal
 </font>


```python
# This is from Techlent notebook
class Solution(object):
    def postorderTraversal(self, root):
        if not root:
            return []
        
        stack = [root]
        rL = []
        
        while stack:
            check = stack.pop()
            rL.append(check.val)
            #note, append left first. 
            if check.left:
                stack.append(check.left)
            if check.right:
                stack.append(check.right)
        # reverse result
        return rL[::-1]
```

```python
# recursive version
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        return  self.postorderTraversal(root.left) + self.postorderTraversal(root.right) + [root.val]
```

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        self.dfs(root, res)
        return res
    
    def dfs(self, root, res):
        if root:
            self.dfs(root.left, res)
            self.dfs(root.right, res)
            res.append(root.val)
```

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:

        traversal, stack = [], [root]
        while stack:
            node = stack.pop()
            if node:
                # pre-order, right first
                traversal.append(node.val)
                stack.append(node.left)
                stack.append(node.right)

        # reverse result
        return traversal[::-1]
```

### <font color='red'>226. Invert Binary Tree
</font>


Given the root of a binary tree, invert the tree, and return its root.
<div></div> 
Input: root = [4,2,7,1,3,6,9]           Output: [4,7,2,9,6,3,1]

```python
# this is from Techlent notebook. 
# use stack to implement DFS preorder traversal.
class Solution(object):
    def invertTree(self, root):
        if not root:
            return
        
        stack = [root]
        while stack:
            check = stack.pop()
            check.left, check.right = check.right, check.left
            if check.left:
                stack.append(check.left)
            if check.right:
                stack.append(check.right)
        return root
```

```python
# this is recursive version. quite intuitive
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:            
            return 
        
        root.left, root.right = root.right, root.left
        root.left = self.invertTree(root.left)
        root.right = self.invertTree(root.right)
        
        return root
```

### <font color='red'>257. Binary Tree Paths</font>


Given the root of a binary tree, return all root-to-leaf paths in any order.
A leaf is a node with no children.
<div></div> 

Input: root = [1,2,3,null,5]      Output: ["1->2->5","1->3"]

```python
# This is from techlent notebook
# We need to remebmer the path for all the node, so when we push root to stack, 
# we keep track of the path so far as well.
class Solution(object):        
    def binaryTreePaths(self, root):
        if not root:
            return []
        res = []
        stack = [(root, "")]
        
        while stack:
            node, path = stack.pop()
            #at the end of a path, a node has no children
            if not node.left and not node.right:
                res.append(path+str(node.val))
            if node.right:
                stack.append((node.right, path+str(node.val)+"->"))
            if node.left:
                stack.append((node.left, path+str(node.val)+"->"))
        return res
```

```python
### recursive version
class Solution:
    def binaryTreePaths(self, root: Optional[TreeNode]) -> List[str]:
        
        if not root: return []
        
        if not root.left and not root.right:
            return [str(root.val)]
    
        treepaths = [str(root.val) +'->' + path for path in self.binaryTreePaths(root.left)]
        treepaths += [str(root.val) + '->' + path for path in self.binaryTreePaths(root.right)]
        
        return treepaths
```

```python
### dfs recursion
class Solution:
    def binaryTreePaths(self, root: Optional[TreeNode]) -> List[str]:
        
        if not root:
            return []
        res = []
        self.dfs(root, "", res)
        return res
    
    def dfs(self, root, ls, res):
        # we can not use if not root here.
        if not root.left and not root.right:
            res.append(ls+str(root.val))
            
        if root.left:
            self.dfs(root.left, ls+str(root.val)+"->", res)
        if root.right:
            self.dfs(root.right, ls+str(root.val)+"->", res)
```

### <font color='red'>515. Find Largest Value in Each Tree Row
</font>


Given the root of a binary tree, return an array of the largest value in each row of the tree (0-indexed).
<div></div> 

Input: root = [1,3,2,5,3,null,9]           Output: [1,3,9]

```python
# This should be best solved by BFS, which is easily coded usig a queue
# BFS with queue is very similar to preorder DFS with a stack
class Solution(object):
    def largestValues(self, root):
        if not root: 
            return []

        from collections import deque
        
        rL= []
        queue = deque([root])
 
        while queue:
            size = len(queue)
            level_max = float('-inf')
            
            #in each for loop, we have added all the nodes from layer below. 
            for i in range(size):
                check = queue.popleft()
                level_max = max(level_max, check.val)
                if check.left:
                    queue.append(check.left)
                if check.right:
                    queue.append(check.right)
                    
            rL.append(level_max)

        return rL
```

```python
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        max_array, self.n = [], 0

        def dfs(cur_node: Optional[TreeNode], cur_row: int):
            if not cur_node:
                return

            if cur_row == self.n:
                self.n += 1
                max_array.append(cur_node.val)

            max_array[cur_row] = max(max_array[cur_row], cur_node.val)

            dfs(cur_node.left, cur_row + 1)
            dfs(cur_node.right, cur_row + 1)

        dfs(root, 0)
        return max_array
```

### <font color='red'>559. Maximum Depth of N-ary Tree </font>


Given a n-ary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

Nary-Tree input serialization is represented in their level order traversal, each group of children is separated by the null value (See examples).
<div></div> 
Input: root = [1,null,3,2,4,null,5,6]       Output: 3

```python
# use DFS    
class Solution(object):
    def maxDepth(self, root):
        if not root: return 0
    
        level = 1
        stack = [(root, level)]
        max_depth = 0
        
        while stack:
            check, level = stack.pop() 
            max_depth = max(max_depth, level)
            for child in check.children:
                stack.append((child, level+1))
        return max_depth
```

```python
#use BFS
class Solution(object):
    def maxDepth(self, root):
        if not root: return 0
    
        level = 1
        from collections import deque
        queue = deque([(root, level)])
        max_depth = 0
            
        while queue:
            check, level = queue.popleft() 
            max_depth = max(max_depth, level)
            for child in check.children:
                queue.append((child, level+1))
        return max_depth
```

```python
# recursive version
class Solution:
    def maxDepth(self, root: 'Node') -> int:
        """
        :type root: Node
        :rtype: int
        """
        if not root: return 0
        
        maxDep = 0        
       
        for children in root.children:
            tmp = self.maxDepth(children)
            maxDep = max(maxDep, tmp)
        
        return maxDep + 1
```

### <font color='red'>589. N-ary Tree Preorder Traversal
</font>


Given the root of an n-ary tree, return the preorder traversal of its nodes' values.

Nary-Tree input serialization is represented in their level order traversal. Each group of children is separated by the null value (See examples)
<div></div> 

Input: root = [1,null,3,2,4,null,5,6]            Output: [1,3,5,6,2,4]

```python
# From techlent notebook. 
class Solution(object):
    def preorder(self, root):
        if not root:
            return []
        
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            res.append(node.val)
            #note, we need to visit the children reversely. 
            for i in range(len(node.children)-1, -1, -1):
                stack.append(node.children[i])
        return res
```

```python
# from MaxMing. This one is also recursion
class Solution:
    def preorder(self, root: 'Node') -> List[int]:
        def dfs(root):
            if root:
                res.append(root.val)
                for child in root.children:
                    dfs(child)    
        res = []
        dfs(root)
        return res
```

```python
# recursive version
class Solution:
    def preorder(self, root: 'Node') -> List[int]:
        if not root: return []
        
        res = []
        res.append(root.val)

        for children in root.children:
            res += self.preorder(children)
                
        return res
```
