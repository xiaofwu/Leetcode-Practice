```python
Leetcode 102 - Binary Tree Level Order Traversal
```

```python
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        if not root: return []
        
        queue = [root]
        ans = []
        
        while queue:
            size = len(queue)
            level = []
            for i in range(size):
                temp = queue.pop(0)
                level.append(temp.val)
                
                if temp.left:
                    queue.append(temp.left)
                if temp.right:
                    queue.append(temp.right)
            ans.append(level)
        return ans
```

```python
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        if not root: return []
        
        stack = []
        ans = []
        
        while stack or root:
            if root: 
                stack.append(root)
                root = root.left
            else:
                temp = stack.pop()
                ans.append(temp.val)
                root = temp.right
        return ans
```

```python
#queue solution
    class Solution(object):
        def isSameTree(self, p, q):
            """
            :type p: TreeNode
            :type q: TreeNode
            :rtype: bool
            """
            # if you understand the stack version, in principle, when you replace the stack with queue, with the rest of the codes the same,
            # you will have the BFS version of this problem.
            # Next, I will show you some optimizations to make the code more concise.
            
            from collections import deque
            # we push p, q together, then you do not need to popleft() twice
            queue = deque([(p, q)])
            
            while queue:
                node1, node2 = queue.popleft()
                if not node1 and not node2:
                    continue
                #not this should be elif , not another if
                elif None in (node1, node2):
                    return False
                else:
                    if node1.val != node2.val:
                        return False
                    queue.append((node1.right, node2.right))
                    queue.append((node1.left, node2.left))
            return True
            
```

```python
class Solution(object):
        def isSameTree(self, p, q):
            """
            :type p: TreeNode
            :type q: TreeNode
            :rtype: bool
            """
            if p and not q:
                return False
            if not p and q:
                return False
            if not p and not q:
                return True
            
            
            stack = [p, q]
            
            while stack:
                p = stack.pop()
                q = stack.pop()
                if p.val != q.val:
                    return False
                
                if p.left and not q.left:
                    return False
                if q.left and not p.left:
                    return False
                if p.left and q.left:
                    stack.append(p.left)
                    stack.append(q.left)
                
                if p.right and not q.right:
                    return False
                if q.right and not p.right:
                    return False
                if p.right and q.right:
                    stack.append(p.right)
                    stack.append(q.right)
            return True
    
```

```python
class Solution(object):
	def isSymmetric(self, root):
		"""
		:type root: TreeNode
		:rtype: bool
		"""
		if not root:
			return True
		def dfs(left,right):
			# 递归的终止条件是两个节点都为空
			# 或者两个节点中有一个为空
			# 或者两个节点的值不相等
			if not (left or right):
				return True
			if not (left and right):
				return False
			if left.val!=right.val:
				return False
			return dfs(left.left,right.right) and dfs(left.right,right.left)
		# 用递归函数，比较左节点，右节点
		return dfs(root.left,root.right)

作者：wang_ni_ma
链接：https://leetcode.cn/problems/symmetric-tree/solution/dong-hua-yan-shi-101-dui-cheng-er-cha-shu-by-user7/
```

```python
class Solution(object):
	def isSymmetric(self, root):
		"""
		:type root: TreeNode
		:rtype: bool
		"""
		if not root or not (root.left or root.right):
			return True
		# 用队列保存节点	
		queue = [root.left,root.right]
		while queue:
			# 从队列中取出两个节点，再比较这两个节点
			left = queue.pop(0)
			right = queue.pop(0)
			# 如果两个节点都为空就继续循环，两者有一个为空就返回false
			if not (left or right):
				continue
			if not (left and right):
				return False
			if left.val!=right.val:
				return False
			# 将左节点的左孩子， 右节点的右孩子放入队列
			queue.append(left.left)
			queue.append(right.right)
			# 将左节点的右孩子，右节点的左孩子放入队列
			queue.append(left.right)
			queue.append(right.left)
		return True

作者：wang_ni_ma
链接：https://leetcode.cn/problems/symmetric-tree/solution/dong-hua-yan-shi-101-dui-cheng-er-cha-shu-by-user7/
```

```python
Leetcode 104 - Maximum Depth of Binary Tree
```

```python
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root: 
            return 0
       
        left = self.maxDepth(root.left)
        
        right = self.maxDepth(root.right)
        
        return max(left,right) + 1
```

```python
class Solution(object):
        def maxDepth(self, root):
            """
            :type root: TreeNode
            :rtype: int
            """
            if not root:
                return 0
            level = 1
            max_depth = 0
    
            #This is the queue version.
            #It is the level order or BFS way to check each level.
            from collections import deque
            queue = deque([])
            queue.append((root, level))
            while queue:
                check, level = queue.popleft()
                max_depth = max(max_depth, level)
                if check.left:
                    queue.append((check.left, level+1))
                if check.right:
                    queue.append((check.right, level+1))
            return max_depth
    
```

```python
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        def dfs(root,ans):
            if not root: return 
            
            ans.append(root.val)
            if root.left:
                dfs(root.left, ans)
            if root.right:
                dfs(root.right,ans)
        ans = []
        dfs(root,ans)
        return ans
        
```

```python
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root:
            return []
        
        stack = [root]
        rL = []
        while stack:
            check = stack.pop()
            rL.append(check.val)
            if check.right:
                stack.append(check.right)
            if check.left:
                stack.append(check.left)
        return rL
```

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        def dfs(root,ans):
            if not root: return 
            
            if root.left:
                dfs(root.left, ans)
            if root.right:
                dfs(root.right,ans)
            ans.append(root.val)
        ans = []
        dfs(root,ans)
        return ans
```

```python
class Solution(object):
    def postorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
		# Therer are other ways to write postorderTraversal
		# This version is to show you that a postorderTraversal is a reverse of preorder.
		# with a minor change, append the left first and then the right nodes.
		
        if not root:
            return []
        
        stack = [root]
        rL = []
        
        while stack:
            check = stack.pop()
            rL.append(check.val)
            #note, append left first. 
            if check.left:
                stack.append(check.left)
            if check.right:
                stack.append(check.right)
        # reverse result
        return rL[::-1]
```

```python
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        
        if not root:
            
            return None
        root.left = self.invertTree(root.left)
        root.right = self.invertTree(root.right)
        root.left, root.right = root.right, root.left
        
       
        
        return root
```

```python
class Solution(object):
    def invertTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        if not root:
            return
        
        stack = [root]

        while stack:
            check = stack.pop()
            check.left, check.right = check.right, check.left
            if check.left:
                stack.append(check.left)
            if check.right:
                stack.append(check.right)
        return root
        
```

```python
class Solution(object):
    def dfs(self, root, curr, rL):
        if not root.left and not root.right:
            #Note the trick here, ''+'a' = 'a'
            rL.append(curr+str(root.val))
            return 
        if root.left:
            self.dfs(root.left, curr+str(root.val)+'->', rL)
            
        if root.right:
            self.dfs(root.right, curr+str(root.val)+'->', rL)       
            
    def binaryTreePaths(self, root):
        """
        :type root: TreeNode
        :rtype: List[str]
        """
        rL = []
        curr = ''
        if root == None:
            return rL
            
        self.dfs(root, curr, rL)
            
        return rL
```

```python
class Solution(object):
        
    def binaryTreePaths(self, root):
        """
        :type root: TreeNode
        :rtype: List[str]
        """

        if not root:
            return []
        res = []
        #We need to remebmer the path for all the node, so when we push root to stack, we keep track of the path so far as well.
        stack = [(root, "")]
        while stack:
            node, path = stack.pop()
            #at the end of a path, a node has no children
            if not node.left and not node.right:
                res.append(path+str(node.val))
            if node.right:
                stack.append((node.right, path+str(node.val)+"->"))
            if node.left:
                stack.append((node.left, path+str(node.val)+"->"))
        return res
```

```python
Leetcode 515 - Find Largest Value in Each Tree Row
```

```python
class Solution(object):
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: 
            return []

        #Using queue, which is BFS
        from collections import deque
        
        rL= []
        queue = deque([root])
 
        while queue:
            size = len(queue)
            level_max = float('-inf')
            #in each for loop, we have added all the nodes in one layer. 
            for i in range(size):
                check = queue.popleft()
                level_max = max(level_max, check.val)
                if check.left:
                    queue.append(check.left)
                if check.right:
                    queue.append(check.right)
                    
            rL.append(level_max)

        return rL
```

```python
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        ans = []
        if not root:
            return ans

        
        queue = [root]
        
        while queue:
            size = len(queue)
            res = []
           
            for _ in range(size):
                a = queue.pop(0)
                res.append(a.val)
                if a.left:
                    queue.append(a.left)
                if a.right:
                    queue.append(a.right)
                result = max(res)
            ans.append(result)
        return ans
```

```python
class Solution:
    def maxDepth(self, root: 'Node') -> int:
        if not root: return 0
        max_depth = 1
        for child in root.children:
            depth = self.maxDepth(child) +  1
            max_depth = max(max_depth,depth)
        return max_depth
```

```python
Leetcode 589 - N-ary Tree Preorder Traversal
```

```python
class Solution:
    def preorder(self, root: 'Node') -> List[int]:
        if not root: return []
        
        ans = [root.val]
        for child in root.children:
            ans.extend(self.preorder(child))
        return ans
```

```python
class Solution(object):
    def preorder(self, root):
        """
        :type root: Node
        :rtype: List[int]
        """
        if not root:
            return []
        
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            res.append(node.val)
            #note, we need to visit the children reversely. 
            for i in range(len(node.children)-1, -1, -1):
                stack.append(node.children[i])
        return res
```
